```python
import pandas as pd
import os

raw_folder = "data/raw"

csv_files = [
    f for f in os.listdir(raw_folder)
    if f.endswith(".csv")
]

print("Files Found:", len(csv_files))
```

```python
for file in csv_files:

    print("\n" + "="*80)
    print("FILE:", file)

    df = pd.read_csv(
        os.path.join(raw_folder,file)
    )

    print("\nShape")
    print(df.shape)

    print("\nData Types")
    print(df.dtypes)

    print("\nHead")
    print(df.head())

    print("\nMissing Values")
    print(df.isnull().sum())
```

```python
for file in csv_files:

    df = pd.read_csv(
        os.path.join(raw_folder,file)
    )

    print("\n" + "="*80)
    print(file)

    print("Duplicate Rows:",
          df.duplicated().sum())

    print("Missing Values:")
    print(df.isnull().sum().sum())
```

```python
import requests
import pandas as pd

schemes = {
    "SBI_Bluechip":119551,
    "ICICI_Bluechip":120503,
    "Nippon_LargeCap":118632,
    "Axis_Bluechip":119092,
    "Kotak_Bluechip":120841
}

for fund, code in schemes.items():

    url = f"https://api.mfapi.in/mf/{code}"

    response = requests.get(url)

    data = response.json()

    df = pd.DataFrame(
        data["data"]
    )

    df.to_csv(
        f"data/raw/{fund}.csv",
        index=False
    )

    print(f"{fund} saved")
```

```python
import os
import shutil

# Get the name of the most recently uploaded file
uploaded_file_name = list(uploaded.keys())[0]

# Source path is the current directory where files.upload() places it
source_path_root = uploaded_file_name

# Destination path in data/raw with the correct name
destination_path_raw = os.path.join("data/raw", "nav_history.csv")

# Check if the source file exists before moving and renaming
if os.path.exists(source_path_root):
    shutil.move(source_path_root, destination_path_raw)
    print(f"Successfully moved and renamed '{uploaded_file_name}' to 'nav_history.csv' in 'data/raw/'.")
else:
    print(f"Error: '{uploaded_file_name}' not found in the root directory. Please ensure the correct file was uploaded.")
```

```python
master = pd.read_csv(
    "data/raw/fund_master.csv"
)

nav = pd.read_csv(
    "data/raw/nav_history.csv"
)

master_codes = set(
    master["Fund Name"]
)

nav_codes = set(
    nav["Fund Name"]
)

missing_codes = (
    master_codes - nav_codes
)

print(
    "Missing Codes:",
    len(missing_codes)
)

print(missing_codes)
```
